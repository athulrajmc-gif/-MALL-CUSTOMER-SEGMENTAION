```python
import pandas as pd
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
print("RAW HEAD:")
print(df.head())
```

```python
print("\nMISSING VALUES:")
print(df.isnull().sum())
```

```python
print("\nDATA TYPES:")
print(df.dtypes)
```

```python
print("\nDUPLICATE ROWS:")
print(df.duplicated().sum())
```

```python
print("\nUNIQUE VALUES IN GENDER:")
print(df['Gender'].unique())
```

```python
print("\nRAW SHAPE:", df.shape)
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned.columns = (
    df_cleaned.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
)
```

```python
print("\nCLEANED COLUMN NAMES:", df_cleaned.columns.tolist())
```

```python
df_cleaned['gender'] = df_cleaned['gender'].str.strip().str.title()
```

```python
df_cleaned['customerid'] = df_cleaned['customerid'].astype(int)
df_cleaned['age'] = df_cleaned['age'].astype(int)
df_cleaned['annual_income_k$'] = df_cleaned['annual_income_k$'].astype(int)
df_cleaned['spending_score_1-100'] = df_cleaned['spending_score_1-100'].astype(int)
```

```python
print("\nCLEANED DATA HEAD:")
print(df_cleaned.head())

print("\nCLEANED DATA TYPES:")
print(df_cleaned.dtypes)

print("\nCLEANED SHAPE:", df_cleaned.shape)
```

```python
df_cleaned = pd.get_dummies(df_cleaned, columns=['gender'], drop_first=True)
```

```python
print("\nCLEANED DATA HEAD:")
print(df_cleaned.head())

print("\nCLEANED DATA TYPES:")
print(df_cleaned.dtypes)

print("\nCLEANED SHAPE:", df_cleaned.shape)
```
